```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(X_train, y_train), (X_test, Y_test) = keras.datasets.mnist.load_data()
X_train
```

Above imported dataset have an array of pixel values that corresponds to 28x28 pixel images

```python
X_train.shape
```

```python
X_train[0].shape #pixel of one image
```

```python
X_train[0] #pixel values of the image # values between 0 - 255
```

```python
y_train #digit in the image 
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[0])
```

Below we are bringing all the values in a range(0-9).
As in neural networks similar values give more accurate results and the conversion is fast.

```python
X_train = X_train/255 
X_test = X_test/255
X_train[0]
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax')) # when you have more than 1 node in output , we use softmax for classification
model.summary()
```

For 1st dense layer Param# is 100,480 which comes from 
784*128 (=100352) + 128 bias = 100,480

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['accuracy']) # metrics for monitoring reports

#The difference b/w is sparse_categorical_crossentropy does not need one hot encoding done while categorical_crossentropy needs one hot encoding

#now we have values ranging from 0-9 so we will use sparse.
```

```python
history = model.fit(X_train, y_train, epochs=25, validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
y_prob
```

```python
y_pred = y_prob.argmax(axis=1)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

The gap between training and validation line means the model is starting to overfit.

Here the model memorized but didn't learn.
